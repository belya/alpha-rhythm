# Какую проблему решаем

Предположим, у нас есть вовлеченность $Eng(t)$ для непрерывного фрагмента игрового процесса длиной $T$. Некоторым $t$ соответствует событие игры (действие)

В состоянии $s_0$ возможны действия $a_0, a_1$. Мы планируем проверить гипотезу о сравнении влияния на вовлеченность каждого из действий в $s_0$ в течение $\tau$ секунд после применения. Мы можем воспользоваться любым из критериев значимости для сравнения мат. ожиданий $x_{s_0, a_0}$ и $x_{s_0, a_1}$.

$$ X_{s, a} = \{<Eng(t)|t \in [t_{start}, t_{start} + \tau]>|t_{start} \in ...\} $$

$$ x_{s,a} = \{AVG(x)|x \in X_{s, a}\} $$

В этом случае статистика двухвыборочного t-критерия Стьюдента может выглядеть так:

$$ t = \frac{AVG(x_{s_0,a_0}) - AVG(x_{s_0,a_1})}{\sqrt{
\frac{s^2_1}{n_1} + \frac{s^2_2}{n_2}
}} $$

Однако в случае, если $s_0$ - редкое состояние, и одно из действий тоже является редким, для достаточной статистической значимости потребуется записать неоправданно большой фрагмент (добавить значения из Joker's Luck). Поэтому предлагается представить оценку $AVG(x_{s, a})$ в виде:

$$ AVG(x_{s, a}) = f_s(s) \circ f_a(a) $$

Таким образом, для проверки гипотезы не придется ждать достаточного количества появлений действия a в состоянии s. 

Мы можем использовать еще меньше состояний, если $f_s$ можно будет выразить с помощью:
- Общих для всех состояний параметров
- Цепочек действий, приводящих к $s$

# Возможные способы реализации

I. Q-функция в терминологии уравнений Беллмана может быть подходящим способом, т.к. $f_a$ в этом случае соответствует повышению вовлеченности после действия, а $f_s$ соответствует $V(s)$. Правда, стоит учесть, что мы будем работать не с исходным $X_{s, a}$, а с дисконтированными значениями
